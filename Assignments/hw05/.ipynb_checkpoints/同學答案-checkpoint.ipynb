```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score

import warnings
warnings.filterwarnings('ignore')

gss_train_df = pd.read_csv('./data/gss_train.csv')
```

```python
gss_train_y = gss_train_df['colrac'].values
gss_train_x = gss_train_df.drop(['colrac'], axis=1).values
```

```python
model_lst = []
logit = LogisticRegression()
model_lst.append(logit)
lda = LinearDiscriminantAnalysis()
model_lst.append(lda)
qda = QuadraticDiscriminantAnalysis()
model_lst.append(qda)
nb = GaussianNB()
model_lst.append(nb)
for k in range(1,11):
    model_lst.append(KNeighborsClassifier(n_neighbors=k))
    
error_lst = [1-np.mean(cross_val_score(model, gss_train_x, gss_train_y, cv=KFold(10), scoring='accuracy')) for model in model_lst]
auc_lst = [np.mean(cross_val_score(model, gss_train_x, gss_train_y, cv=KFold(10), scoring='roc_auc')) for model in model_lst]
```

```python
xrange = np.arange(0, 2*len(model_lst),2)
plt.bar(xrange, error_lst)
plt.ylabel('Error rates')
plt.title('Error rates of the models')
plt.xticks(xrange, ('logit', 'lda', 'qda', 'nb', 'knn\n_1', 'knn\n_2', 'knn\n_3', 'knn\n_4', 'knn\n_5', 'knn\n_6', 'knn\n_7', 'knn\n_8', 'knn\n_9', 'knn\n_10'))
```

```python
xrange = np.arange(0, 2*len(model_lst),2)
plt.bar(xrange, auc_lst)
plt.ylabel('roc_auc')
plt.title('ROC_AUC of the models')
plt.xticks(xrange, ('logit', 'lda', 'qda', 'nb', 'knn\n_1', 'knn\n_2', 'knn\n_3', 'knn\n_4', 'knn\n_5', 'knn\n_6', 'knn\n_7', 'knn\n_8', 'knn\n_9', 'knn\n_10'))
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

gss_test_df = pd.read_csv('./data/gss_test.csv')
gss_test_y = gss_test_df['colrac'].values
gss_test_x = gss_test_df.drop(['colrac'], axis=1).values

logit = LogisticRegression().fit(gss_train_x, gss_train_y)
pred_prob =logit.predict_proba(gss_test_x)[:,1]
pred = logit.predict(gss_test_x)
err = 1-accuracy_score(gss_test_y, pred)
roc_auc = roc_auc_score(gss_test_y, pred_prob)
print('error rate of logistic regression: '+str(err))
print('roc_auc of logistic regression: '+str(roc_auc))
```

```python
from skater.model import InMemoryModel
from skater.core.explanations import Interpretation

model = InMemoryModel(logit.predict_proba, examples = gss_train_x)
interpreter = Interpretation(gss_train_x)
importance = interpreter.feature_importance.feature_importance(model)
```

```python
features = np.array(gss_train_df.drop(['colrac'], axis=1).columns)
plt.figure(figsize=(5,15))
plt.scatter(importance, np.arange(0,len(features)))
_ = plt.yticks(np.arange(0,len(features)), features[list(importance.index)])
_ = plt.title('feature importance')
```

```python
from pycebox.ice import ice, ice_plot

ice_df = ice(gss_train_df.drop(['colrac'], axis=1), 'tolerance', lambda x:logit.predict_proba(x)[:,0], num_grid_points=100)
ice_plot(ice_df, c='dimgray', linewidth=0.3, plot_pdp=True, centered=True)
_ = plt.xlabel('tolerance')
```

```python
from skater.core.global_interpretation.tree_surrogate import TreeSurrogate
gss_train_xnew = gss_train_df[['tolerance', 'colmslm', 'colath', 'science_quiz', 'colmil']].values
logitnew = LogisticRegression().fit(gss_train_xnew, gss_train_y)
model_inst = InMemoryModel(logitnew.predict, examples=gss_train_xnew, feature_names=['tolerance', 'colmslm', 'colath', 'science_quiz', 'colmil'], unique_values=[0, 1],
                      target_names=['0','1'])
# Using the interpreter instance invoke call to the TreeSurrogate
surrogate_explainer =TreeSurrogate(oracle=model_inst, seed=5)
surrogate_explainer.fit(gss_train_xnew, gss_train_y, use_oracle=True, prune='post', scorer_type='default')
surrogate_explainer.plot_global_decisions(colors=['coral', 'lightsteelblue','darkkhaki'])
```
