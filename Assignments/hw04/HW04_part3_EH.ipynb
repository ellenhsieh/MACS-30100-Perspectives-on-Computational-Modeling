```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, cross_val_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score

import warnings
warnings.filterwarnings('ignore')
```

```python
mh_train = pd.read_csv('./data/mh_train.csv').dropna()
mh_test = pd.read_csv('./data/mh_test.csv').dropna()
```

```python
x_train = mh_train[['mhealth_sum', 'age', 'educ', 'black', 'female', 'married', 'inc10']].values
y_train = mh_train[['vote96']].values
x_test = mh_test[['mhealth_sum', 'age', 'educ', 'black', 'female', 'married', 'inc10']].values
y_test = mh_test[['vote96']].values
```

```python
models = []
# Logistic regression model
logit = LogisticRegression()
models.append(logit)

# Linear discriminant model
lda = LinearDiscriminantAnalysis()
models.append(lda)

# Quadratic discriminant model
qda = QuadraticDiscriminantAnalysis()
models.append(qda)

# Naive Bayes
nb = GaussianNB()
models.append(nb)

# K-nearest neighbors with k = 1~10
for k in range(1,11):
    models.append(KNeighborsClassifier(n_neighbors=k))
```

```python
# plot the error rates of the models
error_rates = [1-np.mean(cross_val_score(model, x_train, y_train, 
                                         cv=KFold(10), scoring='accuracy')) for model in models]
bars = ('logit', 'lda', 'qda', 'nb', 'knn_1', 'knn_2', 'knn_3', 'knn_4', 
        'knn_5', 'knn_6', 'knn_7', 'knn_8', 'knn_9', 'knn_10')
x = np.arange(len(bars))
plt.barh(x, error_rates, tick_label=bars)
plt.xlabel('Error rates')
plt.title('Error rate of the Models')
plt.show()
```

```python
# plot AUC of the models
aucs = [np.mean(cross_val_score(model, x_train, y_train, cv=KFold(10), scoring='roc_auc')) for model in models]
plt.barh(x, aucs,tick_label=bars)
plt.xlabel('AUC')
plt.title('AUC of the Models')
plt.show()
```

After analyzing the error rate and AUC of the models, LDA turns out to be the best model because it has the lowest error rate and the highest AUC.

```python
# fit LDA with trianing data
lda = LinearDiscriminantAnalysis().fit(x_train, y_train)
pred_prob = lda.predict_proba(x_test)[:,1]
pred = lda.predict(x_test)
error = 1 - accuracy_score(y_test, pred)
print('Error rate of LDA: {}'.format(error))
```

```python
# calculate AUC of LDA model
auc = roc_auc_score(y_test, lda.predict(x_test))
pred_prob = lda.predict_proba(x_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, pred_prob)

# plot AUC of model
plt.plot(fpr, tpr, label='auc = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], '--', color='red')
plt.xlabel('1 - specificity')
plt.ylabel('sensitivity')
plt.title('LDA')
plt.legend()
print('AUC of LDA:{}'.format(auc))
```
